```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
cereal = pd.read_csv('cereal.csv')
cereal.head()
```

```python
display(cereal.head())
print(cereal.shape)
```

```python
cereal = cereal[['name', 'manufacturer','calories', 'protein', 'fat', 'sodium',
                   'fiber','carbo', 'sugars', 'rating']]
cereal.head()
```

```python
plt.scatter(cereal['sugars'], cereal['rating'])
plt.xlabel('Grams of sugar per serving in Cereal')
plt.ylabel('Rating (Consumer Reports)')
plt.show()
```

```python
plt.scatter(cereal['sugars'], cereal['calories'])
plt.xlabel('Grams of sugar per serving in Cereal')
plt.ylabel('Calories per serving')
plt.show()
```

```python
import seaborn as sns
```

```python
man_info =  cereal[(cereal['sugars'] > 0) & (cereal['rating'] > 0)]
sns.relplot(x='sugars', y='rating', hue="manufacturer", s=200, alpha=.2, data= man_info);
```
